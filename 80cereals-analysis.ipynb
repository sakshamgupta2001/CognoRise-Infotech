```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import data
cereals_data = pd.read_csv('E:\download\cosmic\pro\congo\80 cereals/cereal.csv')
cereals_data.head()
```

```python
# finding out if there is any null values
cereals_data.isnull().sum()
```

```python
# transforming manufacturer from single letter to their full names
cereals_data['mfr'] = cereals_data['mfr'].replace(
    {
        'A':'American Home Food Products',
        'G':'General Mills',
        'K':'Kelloggs',
        'N':'Nabisco',
        'P':'Post',
        'Q':'Quaker Oats',
        'R':'Ralston Purina',
    }
)
cereals_data.head()
```

```python
# group by manufacturer and get sample size
entries_per_manufacturer = cereals_data.groupby('mfr').size()

# plot bar chart
plt.figure(figsize=(14, 7))
plt.title('Entries per manufacturer')
sns.barplot(x=entries_per_manufacturer.index, y=entries_per_manufacturer)
plt.ylabel('Entries')
plt.xlabel('Manufacturer')
```

```python
region = cereals_data['protein'].value_counts()
plt.pie(region,labels=region.keys(),autopct='%.0f%%')
```
